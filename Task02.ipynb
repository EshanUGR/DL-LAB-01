```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv('Au_nanoParticle_dataset.csv')

# 2. Select the 4 features of interest
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
selected_df = df[features]

# 3. Create enhanced pairplot
plt.figure(figsize=(12, 10))
pair_grid = sns.pairplot(selected_df, 
                        diag_kind='hist',  # Histograms on diagonal
                        plot_kws={'alpha': 0.6,  # Scatter plot transparency
                                 'edgecolor': 'k',  # Point edges
                                 's': 40},  # Point size
                        diag_kws={'bins': 15,  # Histogram bins
                                 'color': 'gold',  # Histogram color
                                 'edgecolor': 'black',  # Bar edges
                                 'alpha': 0.7})

# 4. Add correlation coefficients to scatter plots
for i, row in enumerate(pair_grid.axes):
    for j, ax in enumerate(row):
        if i != j:  # Skip diagonal (histograms)
            # Calculate Pearson correlation
            corr = selected_df.iloc[:, j].corr(selected_df.iloc[:, i])
            ax.annotate(f'ρ = {corr:.2f}', 
                       xy=(0.7, 0.9), 
                       xycoords='axes fraction',
                       fontsize=10,
                       bbox=dict(boxstyle='round', 
                               facecolor='white', 
                               alpha=0.8))



plt.suptitle('Pairwise Relationships in Gold Nanopracticel Dataset',
             y=1.02,
             fontsize=14)

plt.tight_layout()
plt.show()
        
```

```python
new_df = data[features]

# Create PairGrid with requested modifications
g = sns.PairGrid(new_df)

# 1. Upper triangle: Bivariate histograms
g.map_upper(sns.histplot) 

# 2. Diagonal: Histograms with KDE
g.map_diag(sns.histplot, kde=True, color='gold', alpha=0.7)

# 3. Lower triangle: Bivariate KDE plots
g.map_lower(sns.kdeplot, fill=True, cmap='Blues', alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```
